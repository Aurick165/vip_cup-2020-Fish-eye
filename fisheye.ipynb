```
import re
import os
import cv2
import random
import pandas as pd
import seaborn as sns
from PIL  import  Image
import matplotlib.pyplot as plt
import numpy as np
```

working with txt files


```
# plot image with box 01_fisheye_day_000000
def show_box(image_id,arr):
    

    img = cv2.imread(train_dir+image_id+'.jpg')
    # opencv always read image as BGR - conversion is must
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    for box in arr:
        
        xm,ym,width,height= box
        xmin=xm-width/2
        xmax=xm+width/2
        ymin=ym-height/2
        ymax=ym+height/2
        # converting string to int

        xmin= int(xmin); ymin= int(ymin)
        xmax= int(xmax) ; ymax= int(ymax)

        

        img= cv2.rectangle(img,(xmin,ymin), (xmax,ymax),(255,0,0),2)
        
        img1 = Image.fromarray(img) # cv2 to PIL
    return img1
```

```
x=2
y=0
z=5

train_dir='/content/984/new_folder/images/'

dir_labels='/content/984/new_folder/labels/01_fisheye_day_000{}{}{}.txt'.format(x,y,z)

image_id='01_fisheye_day_000{}{}{}'.format(x,y,z)
```

```
#getting the labels

arr=[]

with open(dir_labels,'r') as f:
  num_lines = sum(1 for line in open(dir_labels))
  for lines in range(num_lines):
    f_cont=f.readline().split()
    arr.append(f_cont)
arr=np.array(arr)
arr=arr.astype(np.float)
arr = np.delete(arr, 0, 1)
arr=arr*1024
```

```
show_box(image_id,arr)
```
